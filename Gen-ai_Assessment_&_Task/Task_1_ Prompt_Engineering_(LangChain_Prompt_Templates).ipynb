## Import libraries

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate

print("Imports successful")
```

## Task 1: Replace Hardcoded Prompts

```python
def get_basic_explanation_template():
    return PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} in simple terms for beginners"
    )

t1 = get_basic_explanation_template()
print(t1.format(topic="Artificial Intelligence"))
```

## Task 2: Multi-Input Prompt System

```python
def get_multi_input_template():
    return PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} tone"
    )

t2 = get_multi_input_template()

print(t2.format(topic="AI", audience="beginners", tone="friendly"))
print(t2.format(topic="Python", audience="kids", tone="fun"))
print(t2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations Engine

```python
def get_variation_templates():
    return {
        "teaching": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} clearly step by step"
        ),
        "interview": PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 questions about {topic}"
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} as a story"
        )
    }

variations = get_variation_templates()

for key in variations:
    print(key, ":", variations[key].format(topic="Machine Learning"))
```

## Task 4: ChatPromptTemplate System

```python
def get_role_templates():
    return {
        "teacher": "You are a helpful teacher who explains concepts clearly.",
        "interviewer": "You are an interviewer who asks insightful questions.",
        "motivator": "You are a motivational speaker who inspires learning."
    }

def create_chat_prompt(role, topic):
    role_templates = get_role_templates()

    system_template = SystemMessagePromptTemplate.from_template(
        role_templates[role]
    )

    human_template = HumanMessagePromptTemplate.from_template(
        "Tell me about {topic}"
    )

    chat_prompt = ChatPromptTemplate.from_messages(
        [system_template, human_template]
    )

    return chat_prompt.format_messages(topic=topic)

print(create_chat_prompt("teacher", "Neural Networks"))
```

## Task 5: Input Validation Layer

```python
VALID_AUDIENCE = ["beginner", "intermediate", "expert"]
VALID_TONE = ["formal", "casual", "fun"]

def validate_inputs(audience, tone):
    if audience not in VALID_AUDIENCE:
        audience = "beginner"

    if tone not in VALID_TONE:
        tone = "formal"

    return audience, tone

print(validate_inputs("beginner", "fun"))
print(validate_inputs("wrong", "wrong"))
```

## Task 6: Prompt Generator App

```python
def get_final_prompt_template():
    return PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style"
    )


def generate_prompt(topic, audience, tone, style):
    
    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Get final template
    final_template = get_final_prompt_template()

    # Step 3: Generate final prompt
    final_prompt = final_template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )

    return final_prompt

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability Test

```python
def run_reusability_test():
    template = get_multi_input_template()

    test_inputs = [
        {"topic": "AI", "audience": "beginner", "tone": "casual"},
        {"topic": "Python", "audience": "intermediate", "tone": "formal"},
        {"topic": "Deep Learning", "audience": "expert", "tone": "technical"},
        {"topic": "Data Science", "audience": "beginner", "tone": "fun"},
        {"topic": "NLP", "audience": "expert", "tone": "casual"},
    ]

    for case in test_inputs:
        print(template.format(
            topic=case["topic"],
            audience=case["audience"],
            tone=case["tone"]
        ))

run_reusability_test()
```

## Pipeline Flow

```python
def pipeline(topic, audience, tone, style):
    return generate_prompt(topic, audience, tone, style)

print(pipeline("Machine Learning", "beginner", "casual", "teaching"))
```
