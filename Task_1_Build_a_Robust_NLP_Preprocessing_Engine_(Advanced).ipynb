## **Task 1: Conceptual Understanding**

### 1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing, text is treated as data, and even small differences like capitalization can change how words are interpreted. The words "Love" and "love" may look similar to humans, but to a machine, they are considered two completely different tokens. 

For example, "Love" might appear at the beginning of a sentence or be used as a proper noun (like a name), while "love" is usually used as a common verb or noun. If we do not handle this properly, the model may treat them as unrelated words, which can lead to unnecessary duplication and confusion in the dataset. 

---

### 2. What happens if stopwords are not removed?

Stopwords are very common words such as "is", "the", "and", "in", etc., which appear frequently in text but often do not carry significant meaning. If these words are not removed during preprocessing, several issues can arise.

Firstly, the dataset becomes noisy because these words dominate the text without adding useful information. Secondly, it increases computational cost, as the model has to process many unnecessary words. Thirdly, important words may lose their significance because stopwords appear more frequently and can overshadow meaningful terms.However, whether to remove stopwords depends on the task. In some cases, keeping them might be important.

---

### 3. Provide two real-world scenarios where removing stopwords can be harmful.

While removing stopwords is generally useful, there are certain situations where it can negatively affect the outcome.

One important example is sentiment analysis. Consider the sentence: "I am not happy." If we remove the stopword "not", the sentence becomes "I am happy", which completely changes the meaning and leads to incorrect sentiment prediction.


---

### 4. Explain the difference between stemming and lemmatization.

Stemming and lemmatization are both techniques used to reduce words to their base or root form, but they work in different ways.

Stemming is a simpler and faster method that removes suffixes from words using basic rules. For example, words like "running", "runner", and "runs" might all be reduced to "run". However, stemming does not always produce valid words and can sometimes be too aggressive.

Lemmatization, on the other hand, is more advanced and accurate. It uses vocabulary and grammatical rules to convert a word into its correct base form, known as a lemma. For example, "better" would be converted to "good", and "running" would become "run". 

Although lemmatization is slower than stemming, it provides more meaningful and linguistically correct results, which makes it more suitable for real-world NLP applications.

## **Task 2:Build Advanced Preprocessing Function**

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    # 1. Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)
    
    # 2. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 3. Convert to lowercase
    text = text.lower()
    
    # 4. Handle repeated characters (>=3 → 1)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 5. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 6. Tokenization
    tokens = text.split()
    
    # 7. Remove short tokens (≤2 except "no", "not")
    final_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            final_tokens.append(word)
    
    cleaned_sentence = " ".join(final_tokens)
    
    return final_tokens, cleaned_sentence
```

## **Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
import json
output = full_pipeline(test_sentences)
print(json.dumps(output, indent=4))
```
